```python
base_dir = os.getcwd()
file_list = glob(os.path.join(base_dir, '*.csv'))
```

```python
base_dir
```

```python
labels = []
for file in file_list:
    tmp_tb = pd.read_csv(file)
    labels.append(tmp_tb.iloc[:, 1:].values)
```

```python
agg = np.mean(labels, axis=0)
agg_result = np.where(agg > 0.5, 1, 0)
```

```python
agg_result
```
